```python
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df['School_Type'].value_counts().plot(kind='pie', autopct='%.2f%%')
plt.title('School Type')
plt.show()
```

```python
SixHourSleep= df[df['Sleep_Hours'] > 6]
SleepPercent = (len(SixHourSleep) / len(df)) * 100
print(SleepPercent)
```

```python
GenderSleep = SixHourSleep['Gender'].value_counts()
print(GenderSleep)
```

```python
GenderSleep.plot(kind='bar')
plt.title('Bar Graph of Students sleeping more than 6 hours')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
GenderSleep.plot(kind = 'pie', autopct = '%.2f')
plt.title('Pie Chart of Students sleeping more than 6 hours')
plt.show()
```

```python
plt.hist(df['Exam_Score'], bins=50)
plt.title('Exam Scores')
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.show()
```
